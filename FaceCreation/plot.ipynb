```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

```python
files = os.listdir("records")
print(len(files))
```

```python
x = [None]*len(files)
d_loss = [None]*len(files)
d_acc = [None]*len(files)
a_loss = [None]*len(files)
a_acc = [None]*len(files)
for i in range(len(files)):
    f = files[i]
    if f != ".ipynb_checkpoints":
        ss = str.split(f,"_")
        index = int(ss[0])-1
        x[i] = i+1
        d_loss[index] = float(ss[1])
        d_acc[index] = float(ss[2])
        a_loss[index] = float(ss[3])
        a_acc[index] = float(ss[4])
```

```python
plt.plot(x,d_loss,label="d loss")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x,d_acc,label="d acc")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x,a_loss,label="a loss")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x,a_acc,label="a acc")
plt.legend()
plt.grid()
plt.show()
```
